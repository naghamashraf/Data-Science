<h2> B. Single Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
x=df[['ambientIlluminance_Lm']]
y=df[['Output']]

lm.fit(x,y)
yhat = lm.predict(x)

lm.coef_
```

Similar to the coorealation result.<br>
The linear regression between the 'illuminance' and the 'output' infers a negative slope, 
meaning that as the illuminance increases the user tends to be outdoors. <br>
